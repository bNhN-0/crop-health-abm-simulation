# Data Preprocessing for Smart Farming Crop Health Simulation

Prepare the Smart Farming Dataset 2024 for simulation by cleaning the data, scaling numeric features, encoding categorical variables, and generating crop-specific parameter tables.

## Workflow Focus

`raw dataset -> cleaning -> scaling -> scientific feature engineering -> crop parameter lookup table -> prepared dataset`


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

import pandas as pd

from src.data_preprocessing import preprocess_csv
```

```python
data_path = PROJECT_ROOT / "data" / "Smart Farming Dataset 2024.csv"
data_path
```

## Load Raw Dataset

The repository expects the raw dataset in the `data/` folder. The file is not treated as source code; it is an input artifact for the analytics and simulation workflow.


```python
raw_df = pd.read_csv(data_path)
print(raw_df.shape)
raw_df.head()
```

## Run Preprocessing Pipeline

The preprocessing logic is kept in `src/data_preprocessing.py` so the notebook stays concise. The mathematical transformations are the same as the original workflow.


```python
artifacts = preprocess_csv(data_path)
prepared_df = artifacts.prepared_df
crop_params_df = artifacts.crop_params_df
full_feature_df = artifacts.full_feature_df

print(prepared_df.shape)
prepared_df.head()
```

## Crop Parameter Table

Each crop label receives a parameter profile derived from the observed environmental patterns in the dataset. These profiles feed the later aggregation and simulation stages.


```python
crop_params_df.head()
```

## Save Prepared Outputs

These files provide the bridge to the feature aggregation notebook.


```python
prepared_output = PROJECT_ROOT / "data" / "prepared_data_for_dynamic_model.csv"
crop_params_output = PROJECT_ROOT / "data" / "crop_params_table.csv"

prepared_df.to_csv(prepared_output, index=False)
crop_params_df.to_csv(crop_params_output, index=False)

print(f"Saved prepared dataset to: {prepared_output}")
print(f"Saved crop parameters to: {crop_params_output}")
```
